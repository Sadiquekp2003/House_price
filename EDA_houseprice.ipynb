# **OUTLIER DETECTION**

```python
df2=df1.copy()
df2.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot size for all plots
plt.figure(figsize=(16, 12))

# Create a subplot grid for 4 plots (2 rows, 2 columns)
plt.subplot(2, 2, 1)
sns.boxplot(x='area_type', y='price', data=df2, palette="Set2")
plt.title('Price Distribution by Availability Status')
plt.xlabel('Availability Status')
plt.ylabel('Price')

plt.subplot(2, 2, 2)
sns.boxplot(x='area_type', y='total_sqft', data=df2, palette="Set2")
plt.title('Total Sqft Distribution by Availability Status')
plt.xlabel('Availability Status')
plt.ylabel('Total Sqft')

plt.subplot(2, 2, 3)
sns.boxplot(x='area_type', y='bath', data=df2, palette="Set2")
plt.title('Bath Distribution by Availability Status')
plt.xlabel('Availability Status')
plt.ylabel('Number of Bathrooms')

plt.subplot(2, 2, 4)
sns.boxplot(x='area_type', y='balcony', data=df2, palette="Set2")
plt.title('Balcony Distribution by Availability Status')
plt.xlabel('Availability Status')
plt.ylabel('Number of Balconies')

# Adjust the layout to ensure the plots fit nicely
plt.tight_layout()
plt.show()
```

```python
# Define a function to calculate IQR outliers
def detect_iqr_outliers(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (column < lower_bound) | (column > upper_bound)


# Select only numerical columns
#numerical_cols = df2.select_dtypes(include=['number']).columns



# Apply the function to numerical columns
outliers_iqr = df2[[ 'size', 'total_sqft', 'bath',
       'balcony', 'price']].apply(detect_iqr_outliers)

# View outliers
print("Outliers detected using IQR:")
print(outliers_iqr.sum())  # Count of outliers per column

```

```python
# Filter out the rows with any outliers
df2 = df2[~(outliers_iqr.any(axis=1))]
print("Data after removing outliers (IQR):")
print(df2)
```

```python

import numpy as np
from scipy.stats import zscore


# Compute Z-scores
z_scores = df2[[ 'size', 'total_sqft', 'bath',
       'balcony']].apply(zscore)

# Define threshold
threshold = 3

# Identify outliers
outliers_z = (z_scores.abs() > threshold)

# View outliers
print("Outliers detected using Z-Score:")
print(outliers_z.sum())  # Count of outliers per column


```
